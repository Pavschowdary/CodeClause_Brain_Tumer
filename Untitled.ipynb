```python
import requests
from bs4 import BeautifulSoup
import re
import time
from urllib.parse import urljoin, urlparse
import csv
```

```python
# Comprehensive list of UK regions for political contacts
uk_regions = [
"Somerset"
]
```

```python
# Keywords related to political contacts
political_keywords = [
    "politician email",
    "social activist email",
    "government employee email",
    "NGO email",
    "Muslim organization email",
    "Indian organization email",
    "state level minister email",
    "centre level minister email",
    "political party email",
    "parliament member email",
    "congressman email",
    "senator email",
    "assembly member email",
    "mayor email",
    "council member email"
]
```

```python
# Function to fetch Google search results
def fetch_google_search_results(query, start=0):
    url = f"https://www.google.com/search?q={query}&start={start}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers, timeout=10)
    response.raise_for_status()
    return response.text
```

```python
# Function to extract emails from HTML content
def extract_emails_from_html(html):
    email_pattern = r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[com]{3}"
    emails = set(re.findall(email_pattern, html))
    return emails
```

```python
# Function to visit a URL and extract emails
def visit_and_extract_emails(url):
    emails = set()
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        emails.update(extract_emails_from_html(response.text))
    except requests.RequestException as e:
        print(f"Could not extract emails from {url}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {url}: {e}")
    return emails

# Function to get the domain name from URL
def get_domain_name(url):
    parsed_url = urlparse(url)
    return parsed_url.netloc

# Construct search queries for political contacts
search_queries = []
for region in uk_regions:
    for keyword in political_keywords:
        search_queries.append(f"site:nhs.uk {keyword} {region}")

# Store all extracted emails
extracted_emails = []

# Loop through each search query
for search_query in search_queries:
    for page in range(0, 100, 10):  # Adjust the range as needed to control the number of pages searched
        try:
            html = fetch_google_search_results(search_query, start=page)
            soup = BeautifulSoup(html, 'html.parser')

            # Check if there are no results on this page
            if not soup.find_all('div', class_='g'):
                print(f"No more results for {search_query} starting from page {page // 10}")
                break

            # Find all search result links
            links = soup.find_all('a', href=True)

            for link in links:
                href = link['href']
                if "http" in href or "https" in href:
                    print(f"Visiting {href}")
                    emails = visit_and_extract_emails(href)
                    domain_name = get_domain_name(href)
                    for email in emails:
                        extracted_emails.append((domain_name, email, search_query))
                        print(f"Extracted {len(extracted_emails)} emails so far")

            # Add delay to avoid getting blocked
            time.sleep(2)
        except requests.RequestException as e:
            print(f"Failed to fetch search results for {search_query} page {page}: {e}")

# Write extracted emails to CSV file
with open(r'D:\political_related_emails.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Domain Name', 'Email', 'Search Query'])
    csvwriter.writerows(extracted_emails)

print(f"Extracted contact information saved to political_related_emails.csv.")
```
